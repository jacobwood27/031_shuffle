```python
import cv2
import os
import random
```

```python
def prep_frame(f):
    return cv2.resize(f[0:800, 700:1500], (224,224))
```

```python
cap = cv2.VideoCapture("IMG_0976.MOV")
[cap.read() for _ in range(500)]
ret, f = cap.read()

cv2.imwrite("/home/woojac/proj/031_shuffle/post_riffle/sample_full_frame.jpg", f)

cv2.imshow('frame', prep_frame(f))
cv2.waitKey(0)
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture("IMG_0976.MOV")

i = 0

while True:
    
    i+=1

    ret, f = cap.read()
    if not ret:
        print("End of video")
        break

    f = prep_frame(f)
    
    cv2.imshow('frame', f)
    key = cv2.waitKey(0)
    
    if chr(key) in '0cshd':
        cv2.imwrite("/home/woojac/proj/031_shuffle/post_riffle/suit_data/" + chr(key) + "/" + str(i) + ".jpg", f)
    elif chr(key) in 'q':
        break
    else:
        pass

cap.release()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np

fig = plt.figure(figsize=(10., 6.))
grid = ImageGrid(fig, 111, nrows_ncols=(3, 5), axes_pad=0.1, share_all=True)

grid[0].get_yaxis().set_ticks([])
grid[0].get_xaxis().set_ticks([])

grid[0].set_title("C")
grid[1].set_title("D")
grid[2].set_title("H")
grid[3].set_title("S")
grid[4].set_title("0")

ims = []
for _ in range(3):
    for s in "cdhs0":
        im_file = random.choice(os.listdir("post_riffle/suit_data/"+s))
        im = cv2.imread("post_riffle/suit_data/"+s+"/"+im_file)
        ims.append(im[...,::-1])

for ax, im in zip(grid, ims):
    # Iterating over the grid returns the Axes.
    ax.imshow(im)

plt.savefig("suit_verify.png")
```
